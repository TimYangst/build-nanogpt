```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
import numpy as np

# 定义输入范围
x = torch.linspace(-6, 6, 400)

# 1. PyTorch 中精确的 GELU (内部使用 erf 函数)
y_exact = F.gelu(x, approximate='none')

# 2. PyTorch 中使用 tanh 近似的 GELU
y_tanh_approx = F.gelu(x, approximate='tanh')

# 绘图
plt.figure(figsize=(10, 6))
plt.plot(x.numpy(), y_exact.numpy(), label="GELU (erf-based)", linewidth=3, color='blue')
plt.plot(x.numpy(), y_tanh_approx.numpy(), label="GELU (tanh-based)", linestyle='--', linewidth=2, color='red')
plt.title("GELU vs. Tanh")
plt.xlabel("Input (x)")
plt.ylabel("Output")
plt.grid(True)
plt.legend()
plt.show()
```
